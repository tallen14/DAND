# Gather Data

```python
df_archive = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
#Download image-predictions.tsv into the downloads folder.
folder_name = 'downloads'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
r = requests.get(url)
with open(os.path.join(folder_name, url.split('/')[-1]), 'wb') as f:
    f.write(r.content)
```

```python
df_images = pd.read_csv('downloads/image-predictions.tsv',sep='\t')
```

```python
import tweepy
auth = tweepy.OAuthHandler('', '')
auth.set_access_token('', '')
```

```python
from timeit import default_timer as timer
import json

api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

twitter_data={}
tweet_errors=[]
tweet_ids = list(df_archive.tweet_id)

total_start_time = timer()
n=1
for tweet_id in tweet_ids:
    start = timer()
    try:
        tweet_info = api.get_status(tweet_id, tweet_mode='extended')  
        twitter_data[str(tweet_id)] = tweet_info._json
    except Exception as e:
        print("ERROR:" + str(tweet_id) + ": " + str(e))
        tweet_errors.append(tweet_id)  
    end = timer()
    print(n, tweet_id, "%.2f" %(end - start))
    n+=1
                                
with open('tweet_json.txt', 'w') as outfile:
    json.dump(twitter_data, outfile, sort_keys = True, indent=2)

                                
total_end_time = timer() 
total_time = total_end_time - total_start_time
print("Total Download Time: " + str("%.2f" % total_time) + " seconds.")
```

```python
df_tweets = pd.read_json('tweet_json.txt',orient='index')
```

# Assess

The WeRateDogs Twitter archive contains basic tweet data for all 5000+ of their tweets, but not everything. One column the archive does contain though: each tweet's text, which I used to extract rating, dog name, and dog "stage" (i.e. doggo, floofer, pupper, and puppo) to make this Twitter archive "enhanced." Of the 5000+ tweets, I have filtered for tweets with ratings only (there are 2356).

```python
df_archive.head()
```

`df_images` columns:
- **tweet_id** is the last part of the tweet URL after "status/" - (https://twitter.com/dog_rates/status/889531135344209921)
- **p1** is the algorithm's #1 prediction for the image in the tweet
- **p1_conf** is how confident the algorithm is in its #1 prediction
- **p1_dog** is whether or not the #1 prediction is a breed of dog
- **p2** is the algorithm's second most likely prediction
- **p2_conf** is how confident the algorithm is in its #2 prediction
- **p2_dog** is whether or not the #2 prediction is a breed of dog
- etc...

df_images is a table full of image predictions (the top three only) alongside each tweet ID, image URL, and the image number that corresponded to the most confident prediction (numbered 1 to 4 since tweets can have up to four images).

### Visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df_grouped = df_merged.groupby(['day'])['day'].count()
df_grouped = pd.DataFrame(df_grouped)
df_grouped.rename(columns={'day': 'count'}, inplace = True)
df_grouped = df_grouped.reset_index()
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mapping = {day: i for i, day in enumerate(weekdays)}
key = df_grouped['day'].map(mapping)    
df_grouped = df_grouped.iloc[key.argsort()]
df_grouped.plot(kind='bar', x='day',alpha=.6, title='Number of Tweets by Weekday')
plt.ylabel('Tweet Totals')
plt.xlabel('Weekday');
```

```python
df_grouped = df_merged.groupby(['month'])['month'].count()
df_grouped = pd.DataFrame(df_grouped)
df_grouped.rename(columns={'month': 'count'}, inplace = True)
df_grouped = df_grouped.reset_index()
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
mapping = {month: i for i, month in enumerate(months)}
key = df_grouped['month'].map(mapping)    
df_grouped = df_grouped.iloc[key.argsort()]
df_grouped.plot(kind='bar', x='month',alpha=.6, title='Number of Tweets by Month')
plt.ylabel('Tweet Totals')
plt.xlabel('Month');
```
